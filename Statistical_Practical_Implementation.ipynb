Measure Of Central Tendency

1. Mean
2. Median
3. Mode

```python
ages = [23, 24, 32, 45, 12, 43, 67, 45, 32, 56, 32, 120]
```

```python
import numpy as np
print(np.mean(ages))
print(np.median(ages))
```

```python
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
```

```python
import statistics
statistics.mode(ages)
```

```python
import seaborn as sns
sns.boxplot(ages)
```

5 Number Summary (Mean, Median, Mode, 25 percentile, 75 percentile)

```python
import numpy as np
q1, q3 = np.percentile(ages, [25, 75])
```

```python
print(q1, q3)
```

```python
## to check outlier [lower fence - higher fence]
IQR = q3 - q1
lower_fence = q1-1.5*IQR
higher_fence = q3+1.5*IQR
print(lower_fence, higher_fence)
```

Measure of Dispersion
1. Variance
2. Standard Deviation

```python
statistics.variance(ages)
```

```python
np.var(ages, axis=0)
```

```python
def variance(data):
  n=len(ages)
  ## mean of the data
  mean = sum(data)/n
  ## variance
  deviation = [(x - mean) ** 2 for x in data]
  variance = sum(deviation)/(n-1)
 ## variance = sum(deviation)/n
  return variance
```

```python
variance(ages)
```

```python
def variance(data, dof=0):
  n=len(ages)
  ## mean of the data
  mean = sum(data)/n
  ## variance
  deviation = [(x - mean) ** 2 for x in data]
  variance = sum(deviation)/(n-dof)
  return variance
```

```python
variance(ages, dof=1)
```

```python
statistics.pvariance(ages)
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

Check whether Distribution is Normal Distribution

```python
## If you want to check whether feature is gaussian or normal distributed
## Q-Q Plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(sample, dist='norm', plot=pylab)
  plt.show()
```

```python
## create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['sepal_length'])
```

```python
mu, sigma = 3., 1. # mean and standard deviation
sample = np.random.lognormal(mu, sigma, 100)
```

```python
def plot_data_log(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(np.log(sample), dist='norm', plot=pylab)
  plt.show()
```

```python
plot_data_log(sample)
```
